##### Ordinary linear regression (OLS)


```python
import regression
from numpy import *
xArr, yArr = regression.loadDataSet('ex0_regression.txt')
ws = regression.ordi_linear_reg(xArr, yArr)
```

```python
xMat = mat(xArr)
yMat = mat(yArr)
yHat = xMat*ws
```

```python
# plot the curve
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(xMat[:, 1].flatten().A[0], yMat.T[:, 0].flatten().A[0])

xCopy = xMat.copy()
xCopy.sort(0)
yHat = xCopy*ws
ax.plot(xCopy[:, 1], yHat)
```
